```python
# x= 'Cabbage & Buttermilk Curry Recipe'
def name(x):
    l=[]
    for i in x.split(' '):
        if(i == 'Recipe'):
            break
        else:
            l.append(i)
    return " ".join(l)
```

```python
meal['name']= meal['name'].apply(name)
```

```python
from gensim.parsing.preprocessing import remove_stopwords

# new_text = 'Doddapatre Tambuli Recipe is a popular healing, nourishing cold yoghurt curry from Karnataka. Doddapatre, also known as Big Thyme in English, Ajwain leaves in Hindi, Karpooravalli in Tamil, Owa in Marathi and Pani Koorka in Malayalam - is known to have a lot of medicinal values. It has a very distinct smell and flavour to it, which resembles camphor and is known to improve digestion, relieve chest congestion and have overall good health benefits. It also reduces skin related issues when crushed and the juice applied on sensitive areas. Tambuli is a cold yoghurt curry made by grinding a herb, vegetable or spice with coconut and a handful of spices. Sometimes it is also given a tempering which enhances the flavours. Serve this Doddapatre Tambuli along with hot Steamed Rice and ghee as the first course during a South Indian lunch or dinner.'
# new_filtered_text = remove_stopwords(new_text)

meal['description']= meal['description'].apply(lambda x: remove_stopwords(x))
```

```python
def ingredients(text):
    l=[]
    clean= text.replace("\t","").replace("\n","").strip()
    for i in clean:
        if not i.isdigit():
            l.append(i)
    return "".join(l).strip().replace("  ","").replace(" ,",",")
            
```

```python
meal['ingredients']= meal['ingredients'].apply(ingredients)
```

```python
meal.drop(columns='instructions', inplace= True)
'''Instructions are not requried for the recommendation'''
```

```python
meal['ingredients']= meal['ingredients'].str.replace("(","").str.replace(")","").str.replace(" /"," ").str.replace("/ "," ")
```

```python
meal['tags']= meal['description']+ meal['cuisine']+ meal['course']+ meal['diet']+ meal['ingredients']
```

```python
new_meal= meal[['name','tags','image_url']]
```

```python
new_meal= new_meal.reset_index()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features= 5000, stop_words='english')
```

```python
vectors= cv.fit_transform(new_meal['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vectors)
```

```python
def recommend(meal):
    similar_meal= similarity[new_meal[new_meal['name']== meal].index[0]]
    sorted_meal= sorted(list(enumerate(similar_meal)), reverse= True, key= lambda x: x[1])[1:6]
    
    for i in sorted_meal:
        print(new_meal['name'].iloc[i[0]])
```
